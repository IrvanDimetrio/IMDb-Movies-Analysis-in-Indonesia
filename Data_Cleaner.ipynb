```python
import pandas as pd 
import numpy as np 
import re
```

```python
output_file_path = 'D:/Jupyter Notebook Project/Kaggle/IMDB Movies Indonesian/Data/Raw/imdb indonesia star clean.csv'
df = pd.read_csv(output_file_path, encoding='unicode_escape',sep=';')
```

```python
df['Runtime (Minutes)'] = df['Runtime (Minutes)'].str.replace('min', '', regex=True)
df.head()
```

```python
df.info()
```

```python
df['Runtime (Minutes)'] = df['Runtime (Minutes)'].astype('float')
```

```python
df['Runtime (Minutes)'].describe()
```

```python
df['Runtime (Minutes)'] = df['Runtime (Minutes)'].fillna(df['Runtime (Minutes)'].median())
```

```python
df.isnull().sum()
```

```python
df['Runtime (Minutes)'] = df['Runtime (Minutes)'].astype('int')
```

```python
df.head(10)
```

```python
# df['Rating'] = df['Rating'].astype('float')

df['Rating'] = df['Rating'].str.replace(',', '.').astype('float')
```

```python
df.head(10)
```

```python
df.info()
```

```python
# data['Rating'] = data['Rating'].replace(np.nan, 0)

df['Rating'] = df['Rating'].apply(lambda x: x if pd.notnull(x) else np.random.uniform(df['Rating'].min(), df['Rating'].max()))
df['Rating'] = df['Rating'].round(decimals = 1)
```

```python
df.head(10)
```

```python
df['Votes'] = df['Votes'].str.replace(',', '').astype('float')
```

```python
df['Votes'] = df['Votes'].fillna(df['Votes'].interpolate())
```

```python
df.head(10)
```

```python
df['Votes'] = df['Votes'].astype('int')
```

```python
df.tail(10)
```

```python
output_file_path_excel = 'D:/Jupyter Notebook Project/Kaggle/IMDB Movies Indonesian/Data/Raw/imdb indonesia clean.xlsx'
output_file_path_csv = 'D:/Jupyter Notebook Project/Kaggle/IMDB Movies Indonesian/Data/Raw/imdb indonesia clean.csv'
df.to_excel(output_file_path_excel, index=False)
df.to_csv(output_file_path_csv, sep=',', index=False) #csv separator used = ','
```
